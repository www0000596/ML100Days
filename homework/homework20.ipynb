```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize = (6,4),facecolor='black') 

#決定底框
plt.axes([0.1,0.1,.5,.5])
#給定刻度, 若不給定值, 圖的周邊無文字
plt.xticks([]), plt.yticks([])
plt.text(0.1,0.1, 'axes([0.1,0.1,.5,.5])',ha='left',va='center',size=16,alpha=.5)
# 第二層框
plt.axes([0.3,0.3,0.5,0.5])
plt.text(0.1,0.1,'axes([0.3,0.3,.5,.5])',ha = 'left',va = 'center',size = 16,alpha = 0.5)
# 第三層框
plt.axes([0.5,0.5,0.5,0.5])
plt.text(0.1,0.1,'axes([0.5,0.5,.5,.5])',ha = 'left',va = 'center',size = 16,alpha = 0.5)
# 第四層框
plt.axes([0.7,0.7,0.5,0.5])
plt.xticks([]), plt.yticks([])
plt.text(0.1,0.1,'axes([0.7,0.7,.5,.5])',ha = 'left',va = 'center',size = 16,alpha = 0.5)
```

```python
#配置12 組 Bar
n = 12 
x = np.arange(n)
#給定數學運算式
y1 = (1-x/float(n)) * np.random.uniform(0.5,1.0,n)
y2 = (1-x/float(n)) * np.random.uniform(0.5,1.0,n)
#指定上半部繪製區域, 給定 Bar 顏色, 邊界顏色
plt.bar(x, +y1, facecolor='#9999ff', edgecolor='white')
plt.bar(x, -y2, facecolor='lightpink', edgecolor='white')
# +Y 指的是 XY 四象限的第一象限
# -Y 指的是 XY 四象限的第二象限
 #設定繪圖圖示區間
for x,y in zip(x,y1):
    plt.text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')
     #設定Y軸區間
plt.ylim(-1.25,+1.25)
plt.show()
```
