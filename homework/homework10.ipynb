```python
import numpy as np
import pandas as pd
```

```python
#讀取STOCK_DAY_0050_202009.csv、STOCK_DAY_0050_202010.csv
hw_stock1 = pd.read_csv('STOCK_DAY_0050_202009.csv')
hw_stock2 = pd.read_csv('STOCK_DAY_0050_202010.csv')
```

```python
hw_concat = pd.concat([hw_stock1,hw_stock2],axis = 0).set_index('date')
hw_concat
```

```python
#找出open小於close的資料
hw_concat.loc[hw_concat["open"]<hw_concat["close"]]
```
